```python
#[Trading Strategy]Camerilla Pivot Point 
from yahoo_fin import stock_info as si
from pandas_datareader import data as pdr
import pandas as pd
import datetime as dt
import numpy
import talib.abstract as ta
from datetime import datetime
import time
```

```python
stocklist = pd.read_excel('Desktop/Stock_data/Intraday.xlsx')
ticker_list = []
ticker_list = list(stocklist[0])
#ticker_list
# Appending .NS value to the symbol list to make them searched by Yahoo engine
i = 0
while i<len(ticker_list):
    x = ticker_list[i]
    ticker_list[i] = x +'.NS'
    ticker_list[i]
    i=i+1
```

```python
end_date= dt.date.today()
delta = dt.timedelta(1)
start_date=end_date-delta

files=[]

print(start_date)

print(end_date)
```

```python
Name=[]
Time=[]
Indicator =[]
Price=[]
Signals = ["Buy","Sell"]
while True:
    
    Name.append("--")
    Time.append("--")
    Indicator.append("--")
    Price.append("--")
    Signals.append("--")
    for i in ticker_list:
        data = []
        j = 0 
        #data = pdr.get_data_yahoo(i, start="1-1-2021", end="4-1-2021")
        data = pdr.get_data_yahoo(i, start=start_date, end=end_date)
        if len(data)>=2 and len(data)>0:
            #CC - Current Day Close
            #PC - Previous Day Close
            #PH - Previous Day High
            #PL - Previous Day Low
            #PO - Previous Day Open
            CC = data['Close'][1]
            PC = data['Close'][0]
            PH = data['High'][0]
            PL = data['Low'][0]
            PO = data['Open'][0]
            H3 = PC + ((PH - PL)*0.275)
            L3 = PC - ((PH - PL)*0.275)
            H4 = PC + ((PH - PL)*0.55)
            L4 = PC - ((PH - PL)*0.55)


            if (CC<H4 and CC>H3):
            #if (CC>H4):
                #print(i)
                print(i, ":", "Buy Signal")
                Name.append(i)
                Price.append(CC)
                #Fetch current Time
                now = datetime.now()
                current_time = now.strftime("%H:%M:%S")
                Time.append(current_time)
                Indicator.append(Signals[0])

                #d = {'Symbol': [i], 'Signals': [Signal]}
                #df = pd.DataFrame(data=d) 

            if (CC>L4 and CC<L3):
            #if (CC<L4):
                #print(i)
                print(i, ":", "Sell Signal")
                Name.append(i)
                Price.append(CC)
                #Fetch current Time
                now = datetime.now()
                current_time = now.strftime("%H:%M:%S")
                Time.append(current_time)
                Indicator.append(Signals[1])
                #d = {'Symbol': [i], 'Signals': [Signal]}
                #df = pd.DataFrame(data=d)
        else:
            print("Exception")
    d = {'Time': Time,'Stock': Name,'Price': Price,'Signal': Indicator}
    df = pd.DataFrame(d)
    df.to_csv('Desktop/Stock_data/Excel Macro Autorun/Live Feeds.csv')
    time.sleep(600)



   
```
